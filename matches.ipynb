```python
import sqlite3
import pandas as pd
```

```python
def query(sql):
    con = sqlite3.connect("bleague/db.sqlite")
    return pd.read_sql_query(sql, con)
```

```python
query("""
select distinct home from matches
""")
```

```python
df = query("""
select
  *
from
  matches
where
  home in ("横浜BC", "千葉J", "SR渋谷", "川崎", "A東京") and
  (
    home in ("横浜BC", "川崎") or
    away in ("横浜BC", "川崎")
  ) and
  dow in ("sun", "sat")
order by
  year, month, day
""")
df
```

```python
for y, m, d, dow, s, h, a in zip(df.year, df.month, df.day, df.dow, df.start_time, df.home, df.away):
    print("%d-%02d-%02d %s %s〜 %s vs %s" % (y, m, d, "日" if dow == "sun" else "土", s, h, a))
```
